```python
from sklearn.cluster import KMeans
```

```python
x = data.iloc[:,[8,9]].values
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(x)

  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title("The Elbow point graph")
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init="k-means++", random_state=0)
Y = kmeans.fit_predict(x)
print(Y)
```

```python
plt.figure(figsize=[8,8])
plt.scatter(x[Y==0,0], x[Y==0,1], s=50, c='green')
plt.scatter(x[Y==1,0], x[Y==1,1], s=50, c='red')
plt.scatter(x[Y==2,0], x[Y==2,1], s=50, c='blue')
plt.scatter(x[Y==3,0], x[Y==3,1], s=50, c='orange')
plt.scatter(x[Y==4,0], x[Y==4,1], s=50, c='pink')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='centroids')
plt.show()
```

```python
x = data.drop('fake', axis=1)
y = data['fake']
```

```python
print(x.columns)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_scale = scale.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)

y_predict = model.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict) * 100
```

```python
from sklearn.tree import DecisionTreeClassifier
model_t = DecisionTreeClassifier()

model_t.fit(x_train, y_train)

y_pre = model_t.predict(x_test)

print(accuracy_score(y_test ,y_pre) * 100)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(x_train, y_train)

y_pr = model.predict(x_test)
print(accuracy_score(y_test, y_pr) * 100)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(32, activation="relu", input_shape=(x_train.shape[1],)))

model.add(Dense(16, activation="relu"))

model.add(Dense(8, activation="relu"))

model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
history = model.fit(
    x_train,
    y_train,
    epochs=30,
    batch_size=16,
    validation_split=0.2
)
```

```python
accuracy = model.evaluate(x_test, y_test)

print("Model Accuracy:", accuracy)
```

```python
model.save("model/fake_profile.h5")
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
